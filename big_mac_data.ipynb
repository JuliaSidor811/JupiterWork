```python
from pathlib import Path
import pandas as pd
import numpy as np
import os
import boto3
```

```python
path = r'/Users/juliasidor/Desktop/kraje'
given_files = Path(path).glob('*.csv')
```

```python
#złączenie wszystkich plikow *.csv w jednego DataFrame
tmp = list()

for file in given_files:
    df = pd.read_csv(file, index_col=None, header=0)
    df['country'] = file.stem #nazwa pliku bez rozszerzenia
    tmp.append(df)

df = pd.concat(tmp, axis=0, ignore_index=True)
```

```python
df
```

```python
# pozbycie sie wartosci NaN
df.replace(np.NaN,0,inplace=True)
```

```python
df.to_csv('big_mac_data.csv')
```
