```python
from pathlib import Path
import pandas as pd
import numpy as np
import os
import boto3
```

```python
path = r'/Users/juliasidor/Desktop/kraje'
given_files = Path(path).glob('*.csv')
```

```python
#złączenie wszystkich plikow *.csv w jednego DataFrame
tmp = list()

for file in given_files:
    df = pd.read_csv(file, index_col=None, header=0)
    df['country'] = file.stem #nazwa pliku bez rozszerzenia
    tmp.append(df)

df = pd.concat(tmp, axis=0, ignore_index=True)
```

```python
df
```

```python
# pozbycie sie wartosci NaN
df.replace(np.NaN,0,inplace=True)
```

```python
df
```

```python
#filtruje dane aby pokazywaly wylacznie dane z lipca 2021
df_july_21 = df[(df['Date'] >= '2021-07-01') & (df['Date'] <= '2021-07-31')]
df_july_21.isna().sum()
```

```python
df_july_21
```

```python
#obliczanie Big Mac Indexu
july_21 = df_july_21.set_index('country')
#wyznaczam cene za jednego bigmaca w usa
price_bigmac_usa = july_21.loc['united_states', 'local_price']
#wybieram wszystkie kraje zawarte w liscie po za usa (index = 0)
given_countries_all = df_july_21['country'].tolist()
given_countries_all.remove('united_states')

#indeks wyliczam ze wzorum ktory znalazlam tutaj - https://scipython.com/book/chapter-7-matplotlib/problems/p71/the-big-mac-index/
july_21['big_mac_index'] = 0
for country in given_countries_all:
    price_in_usd = july_21.loc[country, 'dollar_price']
    july_21.loc[country, 'big_mac_index'] = (((price_in_usd - price_bigmac_usa) * 100))/ price_bigmac_usa

july_21.sort_values(by=['big_mac_index'], ascending=True) 
```

```python
#zapisuje wynik do csv i dodaje ja do s3
july_21.to_csv('big_mac_index.csv')
s3 = boto3.client("s3")
s3.upload_file(
    Filename = 'big_mac_index.csv',
    Bucket = 'myawsbucket811',
    Key = 'big_mac_index.csv'
)
```
